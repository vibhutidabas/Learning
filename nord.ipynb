```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
l=list(range(1,100001))
d=[]
for i in l:
    d.append(np.random.randint(300,900))
```

```python
df={}
df['SSN']=l
df['Credit_Score']=d
```

```python
df=pd.DataFrame.from_dict(df)
df.head()
```

```python
df_cluster = df[['SSN', 'Credit_Score']]
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_cluster)
```

```python
# for 300 data points
plt.scatter(df_cluster['SSN'], df_cluster['Credit_Score'], c=kmeans.labels_, cmap='viridis',s=10)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.5)
plt.show()
print(centers)
```

```python
# for 1000 data points
plt.scatter(df_cluster['SSN'], df_cluster['Credit_Score'], c=kmeans.labels_, cmap='viridis',s=10)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.5)
plt.show()
print(centers)
```

```python
# for 1 lakh data points
plt.scatter(df_cluster['SSN'], df_cluster['Credit_Score'], c=kmeans.labels_, cmap='viridis',s=0.01)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.5)
plt.show()
print(centers)
```

As the only key factor is Credit Score, I plotted a scatter plot using SSN. 
For 300 points, it is forming clusters reasonably by almost dividing at 500 and 700 points
