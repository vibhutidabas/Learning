```python
import os
from matplotlib.pyplot import imread
import numpy as np
# from PIL import Image
import cv2
from pytesseract import pytesseract,Output
import re
from google.colab.patches import cv2_imshow

def detection():
  img= cv2.imread('/content/drive/MyDrive/text_detect/aadhar/WhatsApp Image 2022-12-29 at 6.52.40 PM.png')
  print('read!')
  d=pytesseract.image_to_data(img, output_type=Output.DICT)

  def sharpen(img):
    kernel3 = np.array([[0, -1,  0],
                    [-1,  5, -1],
                      [0, -1,  0]])
    sharp_img = cv2.filter2D(src=img, ddepth=-1, kernel=kernel3)
    return sharp_img

  def convert_gray(img):
      img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
      return img

  def blur(img, param):
      img=cv2.medianBlur(img,param)
      return img

  def threshold(img):
      img = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
      return img

  # print(img.keys())
  # img=convert_gray(img)
  img1=sharpen(img)
  img=img1
  # img=threshold(img)
  n_boxes =len(d['text'])


  for i in range(n_boxes):
    if int(float(d['conf'][i]))>60:
      (x,y,w,h)=(d['left'][i], d['top'][i], d['width'][i], d['height'][i])
      img=cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0),2)

  print('boxes made')
  cv2_imshow(img)
```
